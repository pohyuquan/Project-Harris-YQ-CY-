```python
%matplotlib inline
import pandas as pd 
import matplotlib.pyplot as plt, numpy as np
```

```python
df = pd.read_csv("va_presidential.csv")
```

```python
democrat = df[["County/City", "Democratic", "Republican"]].groupby('County/City').mean()

my_colors = ['b', 'g']

i = 0 
while i <= 198:
    row = democrat.iloc[i]
    row.plot(kind='bar', title=df["County/City"][i], stacked = True, colormap = 'Accent')
    
    plt.figure()
    i+=1
```
